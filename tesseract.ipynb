```python
def binarize(file,threshold):
    image_to_threshold=Image.open(file)
    oimage=image_to_threshold.convert("L")
    for x in range(oimage.width):
        for y in range(oimage.height):
            if oimage.getpixel((x,y))< threshold:
                oimage.putpixel((x,y),0)
            else:
                oimage.putpixel((x,y),255)
    oimage.show()
    return oimage


for thresh in range(0,257,64):
    print("Trying with threshold+ ",thresh)    
    #display(binarize(img,thresh))
    print("Tesseract result++++++++++++++++++++++++++++++")
    print(pytesseract.image_to_string(binarize(file,thresh)))
    print("Tesseract result end++++++++++++++++++++++")
```

```python

def binarize(image_to_threshold,threshold):
    oimage=image_to_threshold.convert("L")
    for x in range(oimage.width):
        for y in range(oimage.height):
            if oimage.getpixel((x,y))< threshold:
                oimage.putpixel((x,y),0)
            else:
                oimage.putpixel((x,y),255)
    oimage.show()
    return oimage

image=Image.open("storefront.png")
bounding_box=(1040,460,1090,480)
crop_image=image.crop(bounding_box)
newsize=(crop_image.width*8,crop_image.height*8)
for r in [Image.HAMMING, Image.BICUBIC, Image.LANCZOS]:
    resize=crop_image.resize(newsize,r)
    b=binarize(resize,100)
    print(pytesseract.image_to_string(b))
```
